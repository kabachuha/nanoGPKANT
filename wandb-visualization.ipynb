```python
import matplotlib.pyplot as plt 
import csv 

x = [] 
y = []
filename = 'wandbs/gpt-wandb_export_2024-05-06T21 39 31.658+03 00.csv'
  
with open(filename,'r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    
    for row in plots:
        break
      
    for row in plots: 
        x.append(float(row[0]))
        y.append(float(row[1]))

plt.plot(x, y, color = 'b', label = "GPT")

x = [] 
y = []
filename = 'wandbs/gptkant-wandb_export_2024-05-06T21 15 58.593+03 00.csv'
  
with open(filename,'r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    
    for row in plots:
        break
      
    for row in plots: 
        x.append(float(row[0]))
        y.append(float(row[1]))

plt.plot(x, y, color = 'g', label = "GPKANT")

x = [] 
y = []
filename = 'wandbs/gptboosted-wandb_export_2024-05-06T22 00 10.556+03 00.csv'
  
with open(filename,'r') as csvfile: 
    plots = csv.reader(csvfile, delimiter = ',') 
    
    for row in plots:
        break
      
    for row in plots: 
        x.append(float(row[0]))
        y.append(float(row[1]))

plt.plot(x, y, color = 'r', label = "GPT-balanced")
plt.xlabel('Step') 
plt.ylabel('Val Loss') 
plt.title('GPT/GP-KAN-T comparison') 
plt.legend()
plt.show()
```
